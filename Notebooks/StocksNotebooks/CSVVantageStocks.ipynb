```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import timeit

# Reading in the data
daily_adjusted = pd.read_csv('VantageData/daily_adjusted_IBM.csv')
weekly_adjusted = pd.read_csv('VantageData/weekly_adjusted_IBM.csv')
monthly_adjusted = pd.read_csv('VantageData/monthly_adjusted_IBM.csv')
```

```python
# Seaborn Visualizations

# Daily Adjusted
# fig, ax = plt.subplots(1,5)
sns.lineplot(x='timestamp', y='open', data=daily_adjusted)
# sns.lineplot(x='timestamp', y='close', data=daily_adjusted, ax=ax[1])
# sns.lineplot(x='timestamp', y='high', data=daily_adjusted, ax=ax[2])
# sns.lineplot(x='timestamp', y='low', data=daily_adjusted, ax=ax[3])
# sns.lineplot(x='timestamp', y='volume', data=daily_adjusted, ax=ax[4])
```

```python
# Weekly Adjusted
fig, ax = plt.subplots(1,5)
sns.lineplot(x='timestamp', y='open', data=weekly_adjusted, ax=ax[0])
sns.lineplot(x='timestamp', y='close', data=weekly_adjusted, ax=ax[1])
sns.lineplot(x='timestamp', y='high', data=weekly_adjusted, ax=ax[2])
sns.lineplot(x='timestamp', y='low', data=weekly_adjusted, ax=ax[3])
sns.lineplot(x='timestamp', y='volume', data=weekly_adjusted, ax=ax[4])
```

```python
# Monthly Adjusted
fig, ax = plt.subplots(1,5)
sns.lineplot(x='timestamp', y='open', data=monthly_adjusted, ax=ax[0])
sns.lineplot(x='timestamp', y='close', data=monthly_adjusted, ax=ax[1])
sns.lineplot(x='timestamp', y='high', data=monthly_adjusted, ax=ax[2])
sns.lineplot(x='timestamp', y='low', data=monthly_adjusted, ax=ax[3])
sns.lineplot(x='timestamp', y='volume', data=monthly_adjusted, ax=ax[4])
```
